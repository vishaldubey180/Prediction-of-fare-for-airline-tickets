```python
def extract_hour(df,col):
    df[col+'_hour']=df[col].dt.hour
```

## Handling Outlier data 

- Outlier detection and outlier imputation if available in data ...
- Seperate out youe dependent and independent feature ..........

So to know the outliers we basically use

- distribution plot
- Boxplot

```python
def plot(df,col):
    fig,(ax1,ax2)=plt.subplots(2,1)
    sns.distplot(df[col],ax=ax1)
    sns.boxplot(df[col],ax=ax2)
```

```python
plot(df,'Price')
```

So in box plot we can see that there are some outliers after conversation of domain experties they told me that values after 40000 are outliers 
so we will try to imoute them 

```python
df['Price']= np.where(df['Price']>=40000,df['Price'].median(),df['Price'])
```

```python
plot(df,'Price')
```
